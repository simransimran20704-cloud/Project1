```python
plt.figure(figsize=(12,5))
sns.lineplot(x="Date", y="Close", data=df)
plt.xticks(rotation=90)
plt.title("Closing Price Trend Over Time")
plt.show()
```

## Observation
- The closing price of Yes Bank stock shows significant fluctuations over time.
- There is an overall upward trend during some periods, followed by sharp declines, indicating high market volatility.
- The stock price is not constant and varies with time, suggesting that historical trends can be useful for predicting future closing prices.
- No missing values or breaks are visible in the plotted trend, indicating continuous time-series data.

```python
plt.figure(figsize=(12,5))
plt.plot(df['Date'], df['High'], label='High')
plt.plot(df['Date'], df['Low'], label='Low')
plt.xticks(rotation=90)
plt.legend()
plt.title("High vs Low Stock Price")
plt.show()
```

## Observation
- The High and Low stock prices follow a similar trend over time.
- There is a consistent gap between the High and Low prices, representing the daily price range.
- Some periods show a wider gap, indicating higher market volatility.
- Overall, both prices move in the same direction, suggesting a strong relationship between the High and Low stock prices.

```python
plt.figure(figsize=(8,5))
sns.regplot(x='Open', y='Close', data=df)
plt.title("Open vs Close Regression")
plt.show()
```

## Observation
- The scatter plot with the regression line shows a strong positive linear relationship between the Open and Close stock prices.
- Most data points are close to the regression line, indicating a good linear fit.
- As the opening price increases, the closing price also tends to increase.
- This suggests that the Open price is an important predictor of the Close price and is suitable for use in the Linear Regression model.

```python
sns.jointplot(x='Open', y='Close', data=df, kind='reg')
plt.show()
```

## Observation
- The joint plot shows a strong positive linear relationship between the Open and Close stock prices.
- Most data points are concentrated around the regression line, indicating a good linear fit.
- The marginal distributions show that most stock prices are concentrated within a specific range.
- This confirms that the Open price is a significant feature for predicting the Close price.

```python
plt.figure(figsize=(8,5))
plt.hexbin(df['Open'], df['Close'], gridsize=25, cmap='Blues')
plt.colorbar(label='Count')
plt.xlabel("Open")
plt.ylabel("Close")
plt.title("Hexbin Plot")
plt.show()
```

## Observation
- The hexbin plot shows a high concentration of data points along a positive diagonal pattern.
- This indicates a strong positive relationship between the Open and Close stock prices.
- Darker hexagonal bins represent areas with a higher density of observations.
- The distribution suggests that the Open price is strongly associated with the Close price, making it a useful feature for prediction.

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df['Close'], fill=True)
plt.title("KDE Plot of Close Price")
plt.show()
```

## Observation
- he KDE plot represents the distribution pattern of the Yes Bank Close Price values.
- Most of the closing prices are concentrated within a specific price range.
- The peak of the curve indicates the range where closing prices occur most frequently.
- The distribution shows some variation in stock prices over the period.
- Slight skewness in the distribution indicates that extreme high or low closing prices occurred less frequently.
- This analysis helps in understanding the spread and density of the target variable (Close Price) before building the prediction model.
