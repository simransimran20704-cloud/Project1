```python
corelation = df.select_dtypes(include="number").corr()
```

```python
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns, annot=True)
```

## Observation
- The heatmap shows the correlation between the numerical features of the dataset.
- All features (Open, High, Low, and Close) have a very strong positive correlation (0.98–1.00).
- High is highly correlated with Open and Close (≈0.99).
- Low and Close also show an almost perfect positive correlation (≈1.00).
- This indicates that all stock price variables move in a similar direction and are strongly related.

```python
sns.pairplot(df)
```

## Observation
- The pair plot shows the relationship between all numerical features in the dataset.
- The scatter plots indicate a strong positive linear relationship between Open, High, Low, and Close prices.
- The histograms on the diagonal show the distribution of each feature.
- A few points are slightly away from the main cluster, indicating the presence of some extreme values (outliers).
- Overall, the variables are highly correlated, making the dataset suitable for regression-based machine learning models.


```python
sns.relplot(x="Open", y="Close", data=df)
```

## Observation
- The scatter plot shows a strong positive relationship between Open and Close prices.
- As the Open price increases, the Close price also tends to increase.
- Most data points follow an upward trend, indicating a strong correlation.
- A few points are slightly away from the trend, but overall the relationship is strong.

```python
sns.histplot(df["Close"])
```

## Observation:
- The distribution of Close price is positively skewed.

```python
sns.catplot(x="Date", kind="box", data=df)
```

## Observation:
- The box plot shows the distribution of Yes Bank's Close prices.
- The middle line represents the median Close price.
- The box represents the range where most Close price values lie.
- The whiskers show the overall spread of the Close prices.
- This plot helps identify variation and possible outliers in the Close price column.
